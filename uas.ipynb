```python
%%writefile app.py

import streamlit as st
import requests
import torch
import tensorflow as tf
from tensorflow import keras


st.title("Question Answering")
st.text("What would you like to ask??")

@st.cache(allow_output_mutation=True)
def load_model():
  global model
  
  model = torch.load_model('/content/drive/MyDrive/UAS PBA/Dataset/pytorch_model.bin')
  
  return model

with st.spinner('Loading model into memory.....'):
  model = load_model()

text = st.text_input('Enter your question here...')
if text:
  st.write("Response")
  with st.spinner('Searching for answers....'):
    
    prediction = model.predict(text)
    st.write('answer: {}'.format(prediction[0]))
    st.write('title: {}'.format(prediction[1]))
    st.write('paragraph: {}'.format(prediction[2]))
  st.write("")

  
```
